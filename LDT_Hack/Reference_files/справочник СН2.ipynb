```python
path = 'D:\IT\Hackatones\Lider_IT_22\ТЗ\СН-ТСН\СН-2012'

for dirs,folder,files in os.walk(path):
    print('Выбранный каталог: ', dirs)
    print('Вложенные папки: ', folder)
    print('Файлы в папке: ', files)
    
    path_list = []
    for file in files:
        df = pd.read_excel(os.path.join(dirs, file), na_filter=False, skiprows=13, header=None, index_col=None)
        df = df[~df.isin(df.tail(13))].dropna()
        path_list.append(df)
    data = pd.concat(path_list, ignore_index=True)
    
```

```python
# Удалим столбцы с нан
ch = df.dropna(axis=1, how='all')
ch
```

```python
# Удалим пустые и не нужные столбцы 
ch = data.drop({0, 1, 3, 7, 8, 9, 10, 11, 12, 13, 14, 16, 17, 18, 20}, axis=1)
ch
```

Поработаем с наименованиями работ. Уберем дополнительные пункты.

```python
data1 = ch[4].value_counts().reset_index()
data1
```

```python
data1 = data1[(data1[4] < 5)]
data1
```

```python
list_work = list(data1['index'])
list_work.append('')
#list_work
```

```python
df = (
    ch[ch[4].isin(list_work)]
    .reset_index(drop=True)
    #.drop(columns={'index'})
)
df
```

```python
# Уберем Nan на пропуски
df[19] = df[19].replace(np.nan, '')
df.loc[df[6] == 1, 6] = '' 
df
```

```python
df[56] = df[5].astype(str) + df[6].astype(str)
df[1519] = df[15].astype(str) + df[19].astype(str)
df
```

```python
df_final = df.drop(columns={5,6,15,19})
df_final.info()
df_final
```

```python
# Переименуем столбцы
df_final = (
    df_final
    .rename(columns={2:'code', 4:'work', 56:'count', 1519:'total'})
    .query('count != "чел-ч"')
)
df_final = df_final[~(df_final['code'] == "") | ~(df_final['total'] == "")]

# Преобразуем типы
df_final['total'] = df_final['total'].replace('', np.nan).fillna(0).astype('float')

df_final.info()
df_final.head(10)
```

```python
df1 = (
    df_final
    .drop('total', axis=1)
    .reset_index(drop=True)
)
df1.info()
df1.head(3)
```

```python
df1['count'].unique().tolist()
```

```python
# Заменим позиции
df1.loc[df1['count']=='100 м 2', 'count'] = '100 м2'
```

```python
df1 = df1[['code', 'work', 'count']]
```

```python
df2 = (
    df_final[['total']]
    .reset_index(drop=True)
)
df2.info()
df2.head(3)
```

```python
# Добавим строку вниз, а потом удалим первую строку
df2.loc[len(df2.index)] = [0]
df2 = df2.drop(df2.iloc[0]).reset_index(drop=True)
df2.info()
df2.head(3)
```

```python
df_final = (
    df1
    .join(df2, how='left')
    .query('code != ""')
    .reset_index(drop=True)
)
df_final.info()
df_final
```

```python
# Сохраним очищенный справочник
df_final.to_excel('CH_2012_final.xlsx', index=False)
```
