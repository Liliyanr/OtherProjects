# Обработка справочников СН и ТСН.
## Объединение в один с интерграцией из гугл-диска


```python
import pandas as pd
import numpy as np

pd.set_option('display.max_columns', 500)
pd.options.display.max_colwidth=250
```

### Объединяем справочники СН и ТСН

```python
# Загрузка СН в гугл-документы 
url = 'https://docs.google.com/spreadsheets/d/1YttFU741o8RrkLFdNFO3SbSwFi1DyktQXALWjOXVotw/edit?usp=sharing'
url.split('/')
id = url.split('/')[5]
ch = pd.read_excel(f'https://docs.google.com/spreadsheets/d/{id}/export?format=xlsx')

ch.head(3)
```

```python
# Загрузка ТСН в гугл-документы 
url = 'https://docs.google.com/spreadsheets/d/1s-0X2-fb9GZ8dg6oP6DJ2GEBiki8bVx04eJsbb2KQ8A/edit?usp=sharing'
url.split('/')
id = url.split('/')[5]
tch = pd.read_excel(f'https://docs.google.com/spreadsheets/d/{id}/export?format=xlsx')

tch.head(3)
```

```python
ch_tch = pd.concat([ch, tch])
ch_tch.info()
```

```python
ch_tch['count'].unique().tolist()
```

```python
#Сохраним объединенный справочник
ch_tch.to_excel('ch_tch.xlsx', index=False)
```
